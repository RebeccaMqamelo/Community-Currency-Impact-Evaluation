```python
def create_sample(userData, locations):
    
    dfs = []
    
    for location in locations:
        
        userData = userData_new[userData_new.location.str.contains(location, regex=False)]
        sample = userData.loc[(userData.days_active >= 30) & (userData.trades_out_daily * 7 >= 1) & (userData.trades_in_daily * 7 >= 1)]
        women = sample.loc[(sample.gender == 'female')]
        men = sample.loc[(sample.gender == 'male')]
        n_women = len(women)
        n_men = len(men)

        print('Total population in {}: {}'.format(location, len(userData)))
        print('Sample size in {}: {}'.format(location, len(sample)))
        print('Number of women: {}'.format(n_women))
        print('Number of men: {}'.format(n_men))

        def sample_size(population_size, confidence_level, confidence_interval):
            # Adapted from https://github.com/veekaybee/data/blob/master/samplesize.py

            # SUPPORTED CONFIDENCE LEVELS: 50%, 68%, 90%, 95%, and 99%
            confidence_level_constant = [50,.67], [68,.99], [90,1.64], [95,1.96], [99,2.57]

            Z = 0.0
            p = 0.5
            e = confidence_interval/100.0
            N = population_size
            n_0 = 0.0
            n = 0.0

            # LOOP THROUGH SUPPORTED CONFIDENCE LEVELS AND FIND THE NUM STD
            # DEVIATIONS FOR THAT CONFIDENCE LEVEL
            for i in confidence_level_constant:
                if i[0] == confidence_level:
                    Z = i[1]

            if Z == 0.0:
                return -1

            # CALC SAMPLE SIZE
            n_0 = ((Z**2) * p * (1-p)) / (e**2)

            # ADJUST SAMPLE SIZE FOR FINITE POPULATION
            n = n_0 / (1 + ((n_0 - 1) / float(N)) )

            return int(math.ceil(n)) # THE SAMPLE SIZE

        sample_sz_women = sample_size(n_women, 95.0, 5.0)
        sample_sz_men = sample_size(n_men, 95.0, 5.0)
        print('Female sample size for 95% CI:', sample_sz_women)
        print('Male sample size for 95% CI:', sample_sz_men)

        women_trt = pd.DataFrame(random.sample(women['xDAI_blockchain_address'].tolist(), k=sample_sz_women))
        men_trt = pd.DataFrame(random.sample(men['xDAI_blockchain_address'].tolist(), k=sample_sz_men))
        trt = pd.concat([women_trt, men_trt], ignore_index=True)
        trt.reset_index(drop=True, inplace=True)

        print('Total treatment size in {}: {}'.format(location, len(trt)))
        dfs.append(trt)

        sample.insert(0, 'trt_label', 0)

        for i in trt[0]:
            sample.loc[(sample.xDAI_blockchain_address == i), 'trt_label'] = 1

        print(Counter(sample.trt_label))

        def create_csv(df, filename):
            files_present = glob.glob(filename)
            if not files_present:
                df.to_csv(filename)
            else:
                print('WARNING: This file already exists! Overwriting...' )
                os.remove(filename)
                df.to_csv(filename)

        #create_csv(trt, '/Users/rebeccamqamelo/Desktop/Capstone/Rebecca_' + location + '_trt_wallets.csv')
        #create_csv(sample, '/Users/rebeccamqamelo/Desktop/Capstone/Rebecca_' + location + '_sample.csv')
        print('------------------------------')
        
                
    all_treatment_addresses = pd.concat(dfs, ignore_index=True)
    all_treatment_addresses.reset_index(drop=True, inplace=True)
    print('Total number of treated units across all locations:', len(all_treatment_addresses))
    #all_treatment_addresses.to_csv('/Users/rebeccamqamelo/Desktop/Capstone/Rebecca_all_trt_wallets.csv')
```
